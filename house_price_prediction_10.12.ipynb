```python
cols=["MSZoning","Street","LotShape","LandContour","Utilities","LotConfig","LandSlope",
        "Neighborhood","Condition1","Condition2","BldgType","HouseStyle","RoofStyle",
        "RoofMatl","Exterior1st","Exterior2nd","ExterQual","ExterCond","Foundation",
        "BsmtQual","BsmtCond","BsmtExposure","BsmtFinType1","BsmtFinType2","Heating",
        "HeatingQC","CentralAir","Electrical","KitchenQual","Functional","GarageType",
        "GarageFinish","GarageQual","GarageCond","PavedDrive","SaleType","SaleCondition"]
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
for c in cols:
    df[c]=le.fit_transform(df[c].astype(str))
```

```python
df[["LotFrontage","MasVnrArea","GarageYrBlt"]]=df[["LotFrontage","MasVnrArea","GarageYrBlt"]].fillna(df[["LotFrontage","MasVnrArea","GarageYrBlt"]].mean())
```

```python
df1=df1.drop(columns=["Alley","MasVnrType","FireplaceQu","PoolQC","Fence","MiscFeature"])
```

```python
for c in cols:
    df1[c]=le.fit_transform(df1[c].astype(str))
```

```python
df1[["LotFrontage","MasVnrArea","GarageYrBlt"]]=df1[["LotFrontage","MasVnrArea","GarageYrBlt"]].fillna(df1[["LotFrontage","MasVnrArea","GarageYrBlt"]].mean())
```

```python
cols1=["BsmtFinSF1","BsmtFinSF2","BsmtUnfSF","TotalBsmtSF","BsmtFullBath","BsmtHalfBath","GarageCars","GarageArea"]
```

```python
for c in cols1:
    df1[c]=le.fit_transform(df1[c].astype(str))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

```python
x=df.drop("SalePrice",axis=1)
y=df["SalePrice"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
LR=LinearRegression()
```

```python
LR.fit(x_train,y_train)
```

```python
y_pred=LR.predict(x_test)
```

```python
print("R2 Square:",r2_score(y_test,y_pred))
print("MSE:",mean_squared_error(y_test,y_pred))
```

```python
RF=RandomForestRegressor(n_estimators=100, random_state=42)
```

```python
RF.fit(x_train,y_train)
```

```python
y_pred0=RF.predict(x_test)
```

```python
print("R2 Square:",r2_score(y_test,y_pred0))
print("MSE:",mean_squared_error(y_test,y_pred0))
```

```python
print("Intercept:",LR.intercept_)
print("Coefficients:",LR.coef_)
```

```python
train=LR.score(x_train,y_train)
test=LR.score(x_test,y_test)
print(f"Training Accuracy:{train}\nTesting Accuracy:{test}")
```

```python
train1=RF.score(x_train,y_train)
test1=RF.score(x_test,y_test)
print(f"Training Accuracy:{train1}\nTesting Accuracy:{test1}")
```
